# PMF

```python
import numpy as np
import random
import pandas as pd
```

```python
l=[]
for i in range(1000):
    l.append(random.randint(1,6))
```

```python
l[:5]
```

```python
s=(pd.Series(l).value_counts()/pd.Series(l).value_counts().sum()).sort_index()
```

```python
s
```

```python
s.plot(kind='bar')
```

```python
# Double the dies
l1 = []
for i in range(10000):
    x = random.randint(1,6)
    y = random.randint(1,6)
    l1.append(x+y)
```

```python
s1 = pd.Series(l1)
```

```python
prob = (s1.value_counts()/s1.value_counts().sum()).sort_index()
```

```python
prob.plot(kind='bar')
```

```python
cum = np.cumsum(prob)
```

```python
cum.plot(kind='bar')
```
